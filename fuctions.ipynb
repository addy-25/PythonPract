```python
##Variable Length Arguments
##Positional And Keywords Arguments

## Positional Arguments
def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,"aditya")
```

```python
##Keywords arguments

def print_detail(*args,**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_detail(name="adi",age="32",country="india")
```

```python
def print_detail(*args,**kwargs):
    for val in args:
        print(f" Positional Argument:{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_detail(1,2,3,name="adi",age="32",country="india")
```
